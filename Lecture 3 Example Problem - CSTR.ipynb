Starting from the CSTR design equation:
$$V_{CSTR}=\frac{F_{A0}X}{-r_A}$$

We derived our rate law in terms of conversion earlier:
$$-r_A=kC_A=kC_{A0}\left( \frac{1-X}{1+\epsilon X} \right)$$

So now all we need to do is plug in our values and calculate the volume
$$V_{CSTR}=\frac{F_{A0}X}{kC_{A0}\left( \frac{1-X}{1+\epsilon X} \right)}$$
As a reminder:
* k = 0.5 min-1
* CA0 = 2.0 mol/L
* FA0 = 3.6 mol/min
* X = 0.9
* $\epsilon$ = -2/3

```python
#First, define all your variables
k = 0.5 #min-1
CA0 = 2.0 #mol/L
FA0 = 3.6 #mol/min
X = 0.9
epsilon = -2.0/3.0
```

```python
#Next, calculate V_CSTR
V_CSTR = FA0 * X / ( (k*CA0) * (1-X)/(1+epsilon*X))
```

```python
#Display the output
#print(V_CSTR)
roundedV = round(V_CSTR,4)
print(roundedV)
print("The volume of the CSTR is ", '{:.1f}'.format(V_CSTR), " L")
```

### If we compare this to the PFR volume (4.9L) we see once again that the PFR is smaller than the CSTR, which makes sense since the reaction is first order

$$-r_A=kC_A=kC_{A0}\left( \frac{1-X}{1+\epsilon X} \right)$$


$$\frac{F_{A0}}{-r_A}=\frac{F_{A0}}{kC_{A0}}\left( \frac{1+\epsilon X}{1-X} \right)$$

```python
#First use np.linspace to generate a series of X values for which we'll calculate FA0/-rA (Y)
X = np.linspace(0,0.9,100)
```

```python
#Now calcualte FA0/-rA (and call it Y for simplicity)
Y = FA0 / (k*CA0) * (1+epsilon*X)/(1-X)
```

```python
plt.plot(X,Y)
```

```python
#If we wanted to, we could find the PFR volume by taking the area under the curve
from scipy.integrate import trapz
#Calculate the area under the curve up to your desired conversion
V_PFR = trapz(Y,X)
print(V_PFR)

```
