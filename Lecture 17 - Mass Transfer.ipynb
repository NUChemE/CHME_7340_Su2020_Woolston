```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
U_ratio = np.linspace(1, 10, 50)
X1 = np.array([0.2, 0.5, 0.8])
for i in X1:
    X2 = 1-1./np.exp(np.log(1/(1-i))*(U_ratio)**0.5)
    plt.plot(U_ratio, X2, label = 'X1 ='+str(i))
plt.legend()
plt.xlabel('U2/U1')
plt.ylabel('X2')
```
