## Differential method for kinetic parameter estimation
Example 7-2 (same data as 7-1)

Trityl (A) + Methanol (B) -> Products

CB0 = 0.5 M

First order in methanol

Posit the rate law: $$-r_A = kC_A^\alpha C_B^\beta$$

CB0 = 10x CA0, so concentraton is essentially constant:

$$-r_A = k^*C_A^\alpha$$

Constant volume batch reactor: $$\frac{dC_A}{dt}=k^*C_A^\alpha$$

```python
#Load up the data
t = np.array([0, 50, 100, 150, 200, 250, 300])
CA = np.array([0.05, 0.038, 0.0306, 0.0256, 0.0222, 0.0195, 0.0174])
```

```python
#Plot the data
plt.plot(t, CA)
```

```python
#Write a functino that does 3-point numerical derivative 
def numerical_deriv(CA, dt):
    ###TO-DO, this functino should really calculate dt from the given data, and check that it is constant
    n = len(CA)
    dCAdt = [0.0]*n
    for i in range(n):
        #print(i)
        if i == 0: #First data point
            dCAdt[i] = (-3*CA[0]+4*CA[1]-CA[2])/(2*dt)
        elif i == n-1: #Last data point
            dCAdt[i] = 1/(2*dt)*(CA[i-2]-4*CA[i-1]+3*CA[i])    
        else: #All other data points
            dCAdt[i] = 1/(2*dt)*(CA[i+1]-CA[i-1])
            
        np_dCAdt = np.asarray(dCAdt) #Note that further manipulation requires a np array, so easiest to set that here
    return np_dCAdt
```

```python
#Now plot ln(-dCA/dt) vs ln(CA), and calculate parameters
dCAdt = numerical_deriv(CA, 50)
ln_deriv = np.log(-1.0*dCAdt)
ln_CA = np.log(CA)

plt.plot(ln_CA, ln_deriv)
```

```python
from sklearn.linear_model import LinearRegression
x = ln_CA.reshape((-1, 1))
model = LinearRegression().fit(x, ln_deriv)
r_sq = model.score(x, ln_deriv)
slope = model.coef_
intercept = model.intercept_
print(f'Slope = {slope}')
print(f'The reaction is {slope} order in A')
```

## Integral method (with numerical integration and non-linear data fitting)
Example 7-1 in the textbook

Trityl (A) + Methanol (B) -> Products

CB0 = 0.5 M

First order in methanol

Use integral method to confirm reaction is second order in trityl


```python
t = [0, 50, 100, 150, 200, 250, 300]
CA = [0.05, 0.038, 0.0306, 0.0256, 0.0222, 0.0195, 0.0174]
```

Constant volume batch reactor, so:

$$\frac{dC_A}{dt}=-kC_A^2$$

```python
#Define the derivative function
def deriv(CA, t, k):
    dCAdt = -k*CA**2
    return dCAdt
```

```python
#Define a function that integrates the ODE with a guess of k, that can be called by curve_fit
def integrate(t, k):
    sol = odeint(deriv, CA0, t, args=(k,))
    return sol[:,0]
```

```python
#Now run curvefit
popt, pcov = curve_fit(integrate, t, CA, p0 = 0.1)
print(f'k*:\t {popt[0]:.3f} L/mol-min')

fitted_sol = integrate(t, popt[0])
plt.plot(t, CA)
plt.scatter(t, fitted_sol)
CB0 = 0.5 #M
print(f'k:\t {popt[0]/CB0:.3f} (L/mol)^2/min')
```

```python
#In theory, the solver should also be able to tell us the order of the reaction:
def deriv_2(CA, t, k, alpha):
    dCAdt = -k*CA**alpha
    return dCAdt

def integrate_2(t, k, alpha):
    sol = odeint(deriv_2, CA0, t, args=(k,alpha))
    return sol[:,0]

popt, pcov = curve_fit(integrate_2, t, CA, p0 = [0.1, 1.8])
print(f'k*:\t {popt[0]:.3f} L/mol-min')
print(f'alpha:\t {popt[1]:.3f}')

fitted_sol = integrate_2(t, popt[0], popt[1])
plt.plot(t, CA)
plt.scatter(t, fitted_sol)
CB0 = 0.5 #M
#print(f'k:\t {popt[0]/CB0:.3f} (L/mol)^2/min')
```
