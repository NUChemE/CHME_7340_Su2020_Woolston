```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from scipy.integrate import odeint
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
#Load up the data
t = np.array([0, 50, 100, 150, 200, 250, 300])
CA = np.array([0.05, 0.038, 0.0306, 0.0256, 0.0222, 0.0195, 0.0174])
CB0 = 0.5 #M
CA0 = CA[0]
```

## 0. First things first - plot the original data

```python
#Plot the data
plt.scatter(t, CA)
plt.xlabel('time (s)')
plt.ylabel('CA')
plt.title('Original Data')
```

## 1. Integral method (with analytical solution and linearization)
Example 7-1 in the textbook

Trityl (A) + Methanol (B) -> Products
$$A + B \rightarrow C$$

with rate law $-r_A = kC_A^2C_B$

**Problem Statement:** $C_{B0}$ = 0.5 M, and the reaction is first order in methanol (B). Use integral method to confirm reaction is second order in trityl (A)

**Solution** Because $C_{B0}>>C_{A0}, C_{B} \approx C_{B0}$, and we can write the rate law just in terms of $C_A$:

$$\frac{-dC_A}{dt}=k^{'}C_A^2$$

where $k^{'}=C_{B0}k$

Integrating the rate equation gives us

$$\frac{1}{C_A}=k^{'}t+\frac{1}{CA_0}$$

So if we plot $\frac{1}{C_A}$ as a function of $t$ we should have a linear plot

```python
inv = 1/CA
plt.scatter(t, inv)
```

```python
#Fit a line to this:
x = t.reshape((-1, 1))
model = LinearRegression().fit(x, inv)
r_sq = model.score(x, inv)
slope = float(model.coef_)
intercept = model.intercept_
print(f'Slope =\t{slope:.3f}')
print(f'r^2 = \t{r_sq:.5f}')
print(f'k =\t{slope/CB0:.3f}')
```

```python
#Plot the resulting fit
plt.scatter(t, inv)
inv_fit = t*slope+intercept
plt.plot(t, inv_fit)
```

It is **ALWAYS** a good idea to plot your un-transformed fit data to see how well it matches up. Just because it matches in the transformation, doesn't mean it will match the original data well. (NB: You'll see this with the Lineweaver Burk plots on your homework).

```python
CA_fit = 1/inv_fit
plt.scatter(t, CA, label = 'Experimental')
plt.plot(t, CA_fit, label = 'Integral Method - Linearized Fit')
plt.xlabel('time (s)')
plt.ylabel('CA')
plt.title('Fit Arising from Integral Method')
plt.legend()
```

## 2. Integral method (with numerical integration and non-linear data fitting)
Instead of analytically solving the integral, we could do this numerically, and fit the data using least-squares regression. The first part of this (numericaly integration using **ode_int**) is overkill for such a simple rate equatino, but you will often be faced with systems where analytical integration is not possible. The second part (non-linear regression) was covered in the python demo for Lecture 4 (where we fit parameters to the Arrhenius equation). PLease see that notebook for more details on the use of **curve_fit**


```python
#Define the derivative function
def deriv(CA, t, k): #Note here that k represents k'
    dCAdt = -k*CA**2
    return dCAdt
```

```python
#Define a function that integrates the ODE with a guess of k, that can be called by curve_fit
def integrate(t, k):
    sol = odeint(deriv, CA0, t, args=(k,))
    return sol[:,0]
```

```python
#Now run curvefit
popt, pcov = curve_fit(integrate, t, CA, p0 = 0.1)
print(f'k*:\t {popt[0]:.3f} L/mol-min')

fitted_sol = integrate(t, popt[0])
plt.plot(t, CA, label = 'Experimental Data')
plt.scatter(t, fitted_sol, label = 'Integral Method - Nonlinear Fit')
plt.legend()
CB0 = 0.5 #M
print(f'k:\t {popt[0]/CB0:.3f} (L/mol)^2/min')
```

With this method, since the rate law is being integrated numerically, we don't even have to know ahead of time that we expect it to be first order. We can let curve_fit figure that out for us, too!

```python
#Redefine the derivative function to take a passed value of alpha (exponent)
def deriv_2(CA, t, k, alpha):
    dCAdt = -k*CA**alpha
    return dCAdt

def integrate_2(t, k, alpha): #Redefine the integration function to accept alpha from curve_fit
    sol = odeint(deriv_2, CA0, t, args=(k,alpha))
    return sol[:,0]

popt, pcov = curve_fit(integrate_2, t, CA, p0 = [0.1, 3])
print(f'k*:\t {popt[0]:.3f} L/mol-min')
print(f'alpha:\t {popt[1]:.3f}')

fitted_sol = integrate_2(t, popt[0], popt[1])
plt.plot(t, CA)
plt.scatter(t, fitted_sol)
CB0 = 0.5 #M
#print(f'k:\t {popt[0]/CB0:.3f} (L/mol)^2/min')
```

## 3. Differential method for kinetic parameter estimation
To solve this problem with the differential method, we need a function that conveniently produces the numerical differentiation for us. 

**NOTE: I wrote this function below very quickly last night, and haven't completely checked it for errors. It would behove you to check it before using in your homework assignment!**

```python
#Write a function that does 3-point numerical derivative 
def numerical_deriv(CA, dt):
    ###TO-DO, this functino should really calculate dt from the given data (t), and check that it is constant, or use a more generic numerical differentiation algorithm than FOgler's
    n = len(CA)
    dCAdt = [0.0]*n
    for i in range(n):
        #print(i)
        if i == 0: #First data point
            dCAdt[i] = (-3*CA[0]+4*CA[1]-CA[2])/(2*dt)
        elif i == n-1: #Last data point
            dCAdt[i] = 1/(2*dt)*(CA[i-2]-4*CA[i-1]+3*CA[i])    
        else: #All other data points
            dCAdt[i] = 1/(2*dt)*(CA[i+1]-CA[i-1])
            
        np_dCAdt = np.asarray(dCAdt) #Note that further manipulation requires a np array, so easiest to set that here
    return np_dCAdt
```

```python
#Now plot ln(-dCA/dt) vs ln(CA), and calculate parameters
dCAdt = numerical_deriv(CA, 50)
ln_deriv = np.log(-1.0*dCAdt)
ln_CA = np.log(CA)
```

```python
#Linear regression
x = ln_CA.reshape((-1, 1))
model = LinearRegression().fit(x, ln_deriv)
r_sq = model.score(x, ln_deriv)
slope = float(model.coef_)
intercept = model.intercept_
print(f'Slope = {slope:.2f}')
print(f'r^2 = {r_sq:.5f}')
print(f'The reaction is {slope:.2f} order in A')
print(f'k prime = {np.exp(intercept):.3f}')

y_fit = slope*ln_CA+intercept
plt.scatter(ln_CA, ln_deriv)
plt.plot(ln_CA, y_fit)
plt.xlabel('ln CA')
plt.ylabel('ln -dCA/dt')
```
