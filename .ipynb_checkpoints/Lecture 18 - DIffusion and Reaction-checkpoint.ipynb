```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Make plots look a little nicer with larger font
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (5, 3.5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'large',
         'ytick.labelsize':'large'}
pylab.rcParams.update(params)
```

```python
lambda_range = np.linspace(0.01, 1, 50)
thiele = np.array([0.1, 1, 5, 10, 100])
```

```python
plt.figure(figsize = [8, 8])
for i in thiele:
    psi = 1/lambda_range * (np.sinh(i*lambda_range)/np.sinh(i))
    plt.plot(lambda_range, psi, label = r'$\phi^2 = $'+str(i))
plt.legend()
plt.ylabel(r'$\psi$')
plt.xlabel(r'$\lambda$')
```

```python
plt.figure(figsize = [8, 5])
thiele_range = np.linspace(0.1, 50, 50)
eta = 3/(thiele_range**2) * (thiele_range * (np.cosh(thiele_range)/np.sinh(thiele_range)) - 1)
plt.loglog(thiele_range, eta)
plt.xlabel(r'$\phi$')
plt.grid()
plt.ylabel(r'$\eta$')
```
