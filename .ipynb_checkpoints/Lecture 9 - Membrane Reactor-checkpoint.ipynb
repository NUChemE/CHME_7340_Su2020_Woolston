```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
#from scipy.integrate import solve_ivp (If you wanted to use solve_ivp instead)
%matplotlib inline
```

```python
#Parameters that definitely won't change
V_tot = 500 #dm^3
V_range = np.linspace(0, V_tot, 100)

Fi0 = [10., 0., 0.] #mol/min
CT0 = 0.2 #mol/dm^3
Fi0 = [10., 0., 0.] #mol/min
FT0 = np.sum(Fi0)

def initial_params():
    kc = 0.2 #min-1
    kA = 0.7 #min-1
    K = 0.05 #mol/dm^3
    return kc, kA, K

kc_0, kA_0, K_0 = initial_params()
run_solver(kc_0,kA_0,K_0)
run_solver(0,kA_0,K_0)
run_solver(2.0*kc_0,kA_0,K_0)
run_solver(20.0*kc_0,kA_0,K_0)
```

```python
def deriv(current, V, kc, kA, K):
    FA, FB, FC = current
    FT = np.sum([FA, FB, FC])
    rA = -kA * CT0 * ( (FA/FT) - CT0/K * (FB/FT) * (FC/FT) )
    dFAdV = rA
    dFBdV = -rA - kc * CT0 * (FB/FT)
    dFCdV = -rA
    return dFAdV, dFBdV, dFCdV
```

```python
def run_solver(kc, ka, K):
    sol = odeint(deriv, Fi0, V_range, args=(kc, kA, K))
    conversion = (Fi0[0]-sol[-1,0])/Fi0[0]
    np_sol = np.asarray(sol)
    labels = ['FA', 'FB', 'FC']
    plt.figure()
    for col in range(np_sol.shape[1]): 
        plt.plot(V_range, np_sol[:,col], label=labels[col])
    plt.legend()
    print('Conversion of A = ', str(conversion))
```
