```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
```

```python
k = 0.01 #dm^3/mol-min
CA0 = 8 #mol/dm^3
V = 1000 #dm^3
v = 25 #dm^3/min
```

```python
tau = V/v
print(tau)
```

```python
Da2 = k*CA0*tau
print(Da2)
```

```python
X_CSTR = ((2*Da2+1)-(4*Da2+1)**0.5)/(2*Da2)
print(X_CSTR)
```

To see conversion with segregation model in a CSTR, we need to integrate

$$\bar X = \int_{0}^{\infty}X(t)E(t)dt$$
$$\bar X = \int_{0}^{\infty}\frac{kC_{A0}t}{1+kC_{A0}t}\frac{1}{\tau}e^{-t/\tau}dt$$

We'll do this using odeint, and an absurdly high value for final t (not infinity)

```python
def dXbardt(X, t):
    dXdt = k*CA0*t/(1+k*CA0*t)*1/tau*np.exp(-t/tau)
    return dXdt
```

```python
t_range = np.linspace(0, 400, 100)
sol = odeint(dXbardt, 0, t_range)
plt.plot(t_range, sol)
print(max(sol))
```

```python
#Plot the residence time distribution for a CSTR
E = np.exp(-t_range/tau)/tau
plt.plot(t_range, E)
plt.vlines(tau, 0, 0.025, linestyles = 'dashed')
plt.xlabel('t')
plt.ylabel('E(t)')
#plt.xlim(0, 200)
```

Re-do conversion calculation using maximum mixedness

$$\frac{dX}{d\lambda} = \frac{r_A}{C_{A0}}+\frac{E(\lambda)}{1-F(\lambda)}X$$

Since it is easier to integrate forward than backward, we define

$$z = \bar T-\lambda$$

where T is the maximum lambda we need to consider. This allows us to rewrite the equation as:

$$\frac{dX}{dz} = -\frac{r_A}{C_{A0}}-\frac{E(\bar T-z)}{1-F(\bar T-z)}X$$

Which we then integrate from 0 to T

```python
Tbar = 200
def deriv(X,z):
    y = Tbar-z
    F = 1-np.exp(-y/tau)
    E = np.exp(-y/tau)/tau
    dXdz = k*CA0*(1-X)**2-E*X/(1-F)
    return dXdz
```

```python
z_range = np.linspace(0, Tbar, 100)
sol2 = odeint(deriv, 0, z_range)
plt.plot(z_range, sol2)
print(max(sol2))
```
