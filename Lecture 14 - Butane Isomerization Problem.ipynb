```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
```

```python
#Make plots look a little nicer with larger font
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (5, 3.5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'large',
         'ytick.labelsize':'large'}
pylab.rcParams.update(params)
```

```python
#Load constants
H_rxn = -6900 #J/mol butane
R = 8.314 #J/mol-K
EA = 35000#65700 #J/mol
Kc_T2 = 3.03 #Equilibriu constant at 60C
T2 = 60+273.15 #K Where we have data on equilibrium constant
T1 = 360 #K Where we have data on rate constant
T0 = 330 #K Entry into reactor
CA0 = 9.3 #M (kmol/m^3)
k_T1 = 31.1 #1/hr

FT0 = 163000 #mol/hr
theta = np.array([1, 0.0, 0.1/0.9])

FA0 = 0.9*FT0
print(FA0)

cp = np.array([141.0, 141.0, 161.0]) #J/mol-K
summation = np.sum(theta*cp)
```

```python
def vant_hoff(T):
    #Calculate Kc as a function of temperature
    Kc = Kc_T2*np.exp(H_rxn/R*(1/T2-1/T))
    return Kc
```

```python
def conversion_eq(T):
    Kc = vant_hoff(T)
    X_eq = Kc/(1+Kc)
    return X_eq
```

```python
def arrhenius(T):
    k = k_T1*np.exp(EA/R*(1/T1-1/T))
    return k
```

```python
T_range = np.linspace(100, 600, 100)
```

```python
plt.plot(T_range, vant_hoff(T_range), label = 'Kc')
plt.xlabel('Temperature (K)')
plt.ylabel('Kc')
```

```python
plt.plot(T_range, conversion_eq(T_range), label = 'X_eq')
plt.xlabel('Temperature (K)')
plt.ylabel('Equilibrium Conversion')
```

```python
plt.plot(T_range, arrhenius(T_range))
plt.xlabel('Temperature (K)')
plt.ylabel('Rate Constant (1/hr)')
```

```python
def temp(X, T0):
    T = T0-H_rxn*X/summation
    return T
```

```python
def rate(X, T0):
    T = temp(X, T0)
    k = arrhenius(T)
    Kc = vant_hoff(T)
    rate = k*CA0*(1-(1+1/Kc)*X)
    return rate
```

```python
def deriv(X, V, T0):
    dXdV = rate(X, T0)/FA0
    return dXdV
```

```python
V = 3000 #L
V_range = np.linspace(0, V, 100)
sol = odeint(deriv, 0, V_range, args = (T0,))
```

```python
plt.plot(V_range, sol)
plt.xlabel('V (L)')
plt.ylabel('Conversion (X)')
```

```python
plt.plot(V_range, temp(sol, T0))
plt.xlabel('V (L)')
plt.ylabel('Temperature (K)')
plt.xlim([0, 2500])
```

```python
#We can also generate a Levenspiel plot
X_LP = np.linspace(0, 0.65, 50)
plt.plot(X_LP, FA0/rate(X_LP, T0))
plt.xlabel('X')
plt.ylabel(r'$-\frac{F_{A0}}{r_A}$ (L)')
```

```python
plt.plot(V_range, rate(sol,T0))
plt.xlabel('V (L)')
plt.ylabel('Reaction Rate (mol/L-hr)')
```

```python
#Calculate conversion in 3000L reactor with a range of inlet temperatures
T0_range = np.linspace(300, 360, 5)
V_range_new = np.linspace(0, 4000, 100)

for T0i in T0_range:
    sol = odeint(deriv, 0, V_range_new, args = (T0i,))
    plt.plot(V_range_new, sol, label = 'T0 = '+str(T0i) + ' K')
plt.legend(fontsize = 'medium')
plt.xlabel('V (L)')
plt.ylabel('X')
```
