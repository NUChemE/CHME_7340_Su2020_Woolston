# Conversion vs Da for CSTR and PFR
Assuming first order kinetics

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
Da = np.linspace(0.0, 10.0, 100)
X_CSTR = Da / (1. + Da)
X_PFR = 1-np.exp(-Da)
```

```python
plt.figure(1)
plt.plot(Da,X_CSTR,'r', label='CSTR')
plt.plot(Da,X_PFR,'b', label='PFR')
plt.legend()
plt.xlabel('Da')
plt.ylabel('X')
```
