# Levenspiel plots for different reaction orders
This notebook generates Levenspiel plots for reactions with simple first, second, etc. order kinetics

```python
#Import necessary packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

Using algebra, we can show that for the general rate law
$$r_j = kC_j^n$$
The Levenspiel plot can be represented by
$$\frac{F_{j0}}{-r_j}=\left( \frac{F_{j0}}{kC_{j0}^n} \right) \frac{1}{(1-X)^n}$$
We will now plot this for X = 0.01 to 0.8 for n = -1 to 2

```python
X = np.linspace(0.01,0.8,20) #Define the values of X for which to calculate Y
Fj0 = 1.0 #Arbitrary
k = 1.0 #Arbitrary
Cj0 = 2.0 #Arbitrary
for n in [-1, 0, 1, 2]: #Loop over reaction orders and calculate Levenspiel plot, then plot result with labels
    Y = (Fj0 / (k* (Cj0 ** n))) * 1.0/((1-X)**n) #Note that exponents in Python are handled by a**b
    plt.plot(X,Y, label = "n = "+ str(n))
    plt.xlabel("X")
    plt.ylabel("F0/-rj")
    plt.legend()
```
